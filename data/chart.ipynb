```python
import matplotlib.pyplot as plt

websites = ['IELTS Fighter', 'Dol English', 'IELTS Mentor']
num_essays = [100, 107, 2644]

plt.pie(num_essays, labels=websites, autopct='%1.1f%%')
plt.title('Percentage of Essays from Each Website')
plt.show()
```

```python
import matplotlib.pyplot as plt

data_sources = ['Method 1 (Websites)', 'National Geographic', 'ChatGPT']
num_essays = [sum(num_essays), 2312, 5702]

plt.bar(data_sources, num_essays)
plt.xlabel('Data Sources')
plt.ylabel('Number of Essays')
plt.title('Comparison of Essays Generated from Different Data Sources')
plt.show()
```
